# Hydrogen Spectra

The emission spectrum of atomic hydrogen is an important early measurement that led to the formation of Quantum Mechanics. The discreet spectral emission lines occur due to electron undergoing transitions between energy levels. A consequence of CPT symmetry is that if there exists a matter bound state $|n_i>$ with energy eigenvalue $E_n$, there is also an antimatter bound state $|\bar{n_i}>$ with the same eigenvalue $E_n$. Therefore, antimatter bound states will present the same energy spectrum as their matter equivalents, i.e: antihydrogen is expected to exhibit the same spectral lines as hydrogen.Therefore, measuring the energy levels of antihydrogen provides an extremely precise measurement of CPT symmetry, or any unexpected deviations, when compared to the known energy levels for hydrogen.

## Rydberg Formula

For a long period it had been a well established phenomena that spectrum of light emitted by an element is a distinct trait of said element. This could be experienced when burning an element in a flame, or better still, using a prism to to disperse the light to witness the discreet lines for an element. It was Rydberg who determined the first general formula for the hydrogen series in 1888, finding that the spectral lines obey the formula:
    \begin{equation}
        \frac{1}{\lambda} = R \left( \frac{1}{n^2} - \frac{1}{n^{\prime 2}} \right)
    \end{equation}

```python
def rydberg(n, n_prime):
    R = 10973731.6 #in units of m
    
    k = R*(1/(n**2) - 1/(n_prime**2)) 
    lam = 1/k
    return lam

## The Balmer Series, n = 2
lam=[]
plt.figure(figsize=(18,6))
for i in range(3, 9):
    plt.axvline(x=rydberg(2, i)/1E-9, color=rainbow[i-3])
```

where $R$ is the Rydberg constant, and both $n$ and $n^\prime$ are integers. This formula determined that the Lyman, Balmer, and Paschen series where special cases where $n = 1, 2, \& 3$ respectively. The Balmer series was the most well know, as the other series predicted by the Rydberg formula were not in the optical band and therefore more more difficult to observe experimentally; and their later observation provided weight for this interpretation for Hydrogen spectra. 

Whilst the Rydberg formula provides an empirical statement of the Hydrogen atom spectra, it lacks a proper theoretical interpretation. The Bohr model of the atom delves deeper into providing an explanation for atomic spectra, whilst also laying out groundwork for some principles of quantum mechanics.

## Bohr

The experiments by Rutherford had previously determined the structure of the atom: an extremely small and dense positively charged nucleus, surrounded by a cloud of negatively electrons. The electromagnetic force which holds electrons in place also has a $1/r^2$ dependence, leading to the interpretation that electrons orbit the nucleus in a similar guise to planets orbiting the sun in the solar-system. The hydrogen atom is the most simple of cases: a single electron orbiting a single proton under the influence of the Coulomb force. However, the hydrogen spectra hints at something more peculiar - hydrogen atoms only emit light of particular wavelengths, an early suggestion of certain quantisation.  Neils Bohr came up with the novel idea of quantisation of electron orbit as a means to explain the spectrum - this radical idea represents one of the first steps towards quantum mechanics.

The Bohr model assumes the electron orbits the nucleus in a circle, similar to planetary orbit such that the centripetal acceleration is in balance with the Coulomb force, i.e: 

\begin{equation}
    F = \frac{m_e v}{r} = \frac{e^2}{4\pi \epsilon_0 r^2}
\end{equation}

The energy levels are a sum of kinetic and potential energies (negative), $E = \frac{1}{2} m_e v^2 -  \frac{e^2}{4\pi \epsilon_0 r}$, to which we can substitute in the equation above to give

\begin{equation}
   E =  - \frac{e^2/4\pi \epsilon_0}{2r}
\end{equation}

This is of no huge consequence, however Bohr makes a radical assumption: only certain orbits are allowed. This produces fixed discreet energy levels for the electron. The electron can loose or gain energy by making quantum "jumps" between the allowed orbits or energy levels. The emission lines of the hydrogen spectra are therefore emitted due to electrons going from one energy level to a lower energy level, emitting light of a wavelength characteristic of the difference in energy between the two levels.

Bohr determined the allowed orbits as the ones which quantise angular momentum in integer multiples of $\hbar$, i.e $m_e v r = n \hbar$. This in turn gives the allowed quantised orbital radii $r_n = a_0 n^2$. Thus, this can be substituted in give the famous Bohr formula for electron energy levels:

\begin{equation}
   E =  - \frac{e^2/4\pi \epsilon_0}{2a_0}\frac{1}{n^2}
\end{equation}

This equation agrees very closely to the observed hydrogen spectrum described by Rydberg, whilst also providing a more detailed theoretical understanding. This interpretation can also be extended to heavier atoms, by exchanging the electron charge $e$ for nuclear charge $Ze$, and exchanging the electron mass $m_e$ for the modified reduced mass of electron $\frac{m_e M}{m_e + M}$ where $M$ is the mass of the nucleus.

## Schrodinger

Whilst the Bohr model demonstrated some success in providing a theoretical underpinning for the Rydberg formula by introducing some quantum aspects, it is still an incomplete oversimplification of the hydrogen spectra. To get a more precise understanding of the phenomena, the Bohr model is superseded by more the accurate quantum mechanical methods involving the wavefunction of the electron. This new framework computes the equivalent energy levels as the Bohr model, whilst providing more information that can be used to determine further insights: such as the transition rates between levels.

The solution of the Schrodinger equation for an electron of mass $m_e$ in a spherically symmetric potential models the electron in the hydrogen atom

\begin{equation}
    \left\{\frac{-\hbar^{2}}{2 m_{e}} \nabla^{2}+V(r)\right\} \psi=E \psi
\end{equation}

where the Laplace operator $\nabla^2$ is equal to

\begin{equation}
\begin{split}
        \nabla^2 \psi & = \frac{1}{r^{2}} \frac{\partial}{\partial r}\left(r^{2} \frac{\partial \psi}{\partial r}\right)+\frac{1}{r^{2} \sin \theta} \frac{\partial}{\partial \theta}\left(\sin \theta \frac{\partial \psi}{\partial \theta}\right)+\frac{1}{r^{2} \sin ^{2} \theta} \frac{\partial^{2} \psi}{\partial \phi^{2}} \\
\end{split}
\end{equation}

We can now follow the standard procedure of separation of variables to solve the partial differential equations. We define the wavefunction as a product of functions $\psi = R(r)Y(\theta, \phi)$. With a bit of rearranging we can get

\begin{equation}
     \frac{1}{R(r)} \frac{\partial}{\partial r}\left[r^{2} \frac{\partial R(r)}{\partial r}\right]-\frac{2 m_{e}}{\hbar^{2}} r^{2}[V(r)-E]=\frac{1}{Y(\theta, \phi)} \hat{\vec{l}}^{2} Y(\theta, \phi)
\end{equation}
Where $\hat{\vec{l}}^{2}$ is equal too the angular part of the laplace operator:
 \begin{equation}
\hat{l}^{2}=-\left\{\frac{1}{\sin \theta} \frac{\partial}{\partial \theta}\left(\sin \theta \frac{\partial}{\partial \theta}\right)+\frac{1}{\sin ^{2} \theta} \frac{\partial^{2}}{\partial \phi^{2}}\right\}
\end{equation}

The equation of motion is now split into angular and radial parts on either side of the equals, thus both being equal to a constant, which we shall call $b$. 

#### A brief consideration of angular momentum

Before going further lets take a moment to consider the angular momentum operator $\hat{\vec{L}}$. We can write the angular momentum operator $\hat{\vec{L}}$ in spherical polar coordinates, its messy but its definaiition herere comes form the classical definition! Remember, angualar momentum is a vector, so the operator will be a vector of operators:

\begin{equation}
\hat{\vec{L}} = (\hat{L_x}, \hat{L_y}, \hat{L_z}) = -i \hbar\left(\sin \phi \frac{\partial}{\partial \theta}+\cot \theta \cos \phi \frac{\partial}{\partial \phi},-\cos \phi \frac{\partial}{\partial \theta}+\cot \theta \sin \phi \frac{\partial}{\partial \phi},-\frac{\partial}{\partial \phi}\right) 
\end{equation}

If we use the above expression to work out $\hat{L^2} = \hat{L_x}^2+\hat{L_y}^2+\hat{L_z}^2$ we find that this corresponds quite nicely with the angualr part of the laplace operator, such that $\hbar \hat{L^2} = \hat{l^2}$! Going forward it may also be worth considering the commuation relations of angular momentum. $\hat{\vec{L}}^{2}$ commutes with the components but the components don't commute with eachother, thus meaning they cannot be known with precision simulatanously.

\begin{equation}
{\left[\hat{L}_{x}, \hat{L}_{y}\right]=\hat{L}_{x} \hat{L}_{y}-\hat{L}_{y} \hat{L}_{x}=i \hbar \hat{L}_{z}} \\
\end{equation}
\begin{equation}
{\left[\hat{L}_{y}, \hat{L}_{z}\right]=i \hbar \hat{L}_{x}} \\
\end{equation}
\begin{equation}
{\left[\hat{L}_{z}, \hat{L}_{x}\right]=i \hbar \hat{L}_{y}} \\
\end{equation}
\begin{equation}
\left[\hat{L}_{x}, \hat{\vec{L}}^{2}\right]=\left[\hat{L}_{y}, \hat{\vec{L}}^{2}\right]=\left[\hat{L}_{z}, \hat{\vec{L}}^{2}\right]=0
\end{equation}

$\hat{L^2}$ and $\hat{L_z}$ have simultaneous eigenfunctions $Y$ with eigenvalues 

\begin{equation}
\begin{array}{l}
\hat{\vec{L}}^{2} Y=\hbar^{2} l(l+1) Y \\
\hat{L}_{z} Y=\hbar m_{l} Y .
\end{array}
\end{equation}

Here, $l$ = (0,1,2, ‚Ä¶ ) and $m_l$ = (‚àíùëô, ‚àíùëô + 1, ùëô ‚àí 1, ùëô) are quantum numbers. We are anticipating the
result that these eigenfunctions will be solutions to the angular part of the Schr√∂dinger equation.
We shall see that this will be exactly the case, and when we solve the radial equation, we will find
that $l$ cannot exceed $n$ ‚àí 1. We will also give these new quantum numbers their proper names. 

#### The Angular Schrodinger equation
Dealing with the angular component first, and splitting $Y(\theta, \phi) = \Theta(\theta)\Phi(\phi)$ we can get again find that it is seperable, and equal to a constant which we will call $m_l$

\begin{equation}
    \begin{split}
    -b &=\frac{1}{ Y \sin \theta} \frac{\partial}{\partial \theta}\left(\sin \theta \frac{\partial Y}{\partial \theta}\right) + \frac{1}{Y \sin ^{2} \theta} \frac{\partial^{2} Y}{\partial \phi^{2}} \\
    - \frac{1}{\Phi} \frac{\partial^{2} \Phi}{\partial \phi^{2}} &= \frac{ \sin \theta}{ \Theta} \frac{\partial}{\partial \theta}\left(\sin \theta \frac{\partial \Theta}{\partial \theta}\right) + b \sin^2\theta  = const. = m_l^2
    \end{split}
\end{equation}

So, the effect of all this work is that the Schr√∂dinger equation for a hydrogen-like atom, which is a three dimensional partial differential equation is seperable and can be decomposed into three ordinary differential 
equations:

\begin{equation}
\begin{array}{r}
\frac{1}{r^{2}} \frac{d}{d r}\left[r^{2} \frac{d R(r)}{d r}\right]-\frac{2 m_{e}}{\hbar^{2}}[V(r)-E] R(r)-\frac{l(l+1)}{r^{2}} R(r)=0 \\
-\frac{1}{\sin \theta} \frac{d}{d \theta}\left(\sin \theta \frac{d \Theta(\theta)}{d \theta}\right)-l(l+1) \sin ^{2} \theta \Theta(\theta)+\frac{m_{l}^{2}}{\sin ^{2} \theta} \Theta(\theta)=0 \\
\frac{d^{2} \Phi(\phi)}{d \phi^{2}}+m_{l}^{2} \Phi(\phi)=0
\end{array}
\end{equation}

#### $\Phi$ solution
The $\Phi$ equation is a simple differential equation, with solution $\Phi(\phi) = Ae^{im\phi} + Be^{-im\phi}$, and the periodic boundary condition $\Phi(\phi) = \Phi(\phi + 2\pi)$ due to the geometry of the system: this restricts $m_l$ to integer values. This is our new quantum number - the magnetic quantum number. Interestingly if we look at our $\Phi(\phi)$ solution, we find that it is an eigenfunctions of $\hat{L_z}$ such that $\widehat{L}_{\mathrm{z}} \Phi(\phi)=-i h \frac{\partial}{\partial \phi} \Phi(\phi)=\hbar m_{l} \Phi(\phi)$ which we stated above.

#### $\Theta$ solution
The $\Theta$ equation is more complicated, so we make a substitution: $w=\cos(\theta)$, $w'=-\sin(\theta)$ and $w^2 = 1 - \sin ^2(\theta)$
\begin{equation}
-\frac{1}{\sin \theta} \frac{d}{d \theta}\left(\sin \theta \frac{d \Theta}{d \theta}\right)-l(l+1) \sin ^{2} \theta \Theta+\frac{m_{l}^{2}}{\sin ^{2} \theta} \Theta=0
\end{equation}
Becomes
\begin{equation}
\frac{d}{d w}\left[\left(1-w^{2}\right) \frac{d \Theta}{d w}\right]-\frac{m_{l}^{2}}{1-w^{2}} \Theta+l(l+1) \Theta=0
\end{equation}
\begin{equation}
\left(1-w^{2}\right) \frac{d^{2} \Theta}{d w^{2}}-2 w \frac{d \Theta}{d w}+\left(l(l+1)-\frac{m_{l}^{2}}{1-w^{2}}\right) \Theta=0
\end{equation}

if $l$ is an integer then this equation is exactly equal to Legendre‚Äôs equation. The Legendre equation has well known solutions called the Legendre polynomials: $\Theta(\cos \theta)=P_{l}^{\left|m_{l}\right|}(\cos \theta)$

#### Spherical Harmonics 
By combining our $\Phi$ and $\Theta$ solutions we can get out our total solution for $Y_{l, m}$ which depend on $l$ and $m_l$

\begin{equation}
\begin{aligned}
Y_{0,0} &=\sqrt{\frac{1}{4 \pi}} \\
Y_{1,0} &=\sqrt{\frac{3}{4 \pi}} \cos \theta \\
Y_{1,+1} &=\mp \sqrt{\frac{3}{8 \pi}} \sin \theta \mathrm{e}^{+i \phi} \\
Y_{2,0} &=\sqrt{\frac{5}{16 \pi}}\left(3 \cos ^{2} \theta-1\right) \\
Y_{2,+1} &=\mp \sqrt{\frac{15}{8 \pi}} \sin \theta \cos \theta \mathrm{e}^{+i \phi} \\
Y_{2,+2} &=\sqrt{\frac{15}{32 \pi}} \sin ^{2} \theta \mathrm{e}^{+2 i \phi} \\
\text { Normalisation: } & \int_{0}^{2 \pi} \int_{0}^{\pi}\left|Y_{l, m}\right|^{2} \sin \theta \mathrm{d} \theta \mathrm{d} \phi=1
\end{aligned}
\end{equation}

### The Radial Equation
The third part of the schrodinger equation which needs to be solved is the radial part, making use of substitution $u(r)= r R(r)$ we can rearrange:
\begin{equation}
\frac{1}{r^{2}} \frac{d}{d r}\left[r^{2} \frac{d R(r)}{d r}\right]-\frac{2 m_{e}}{\hbar^{2}}[V(r)-E] R(r)-\frac{l(l+1)}{r^{2}} R(r)=0 
\end{equation}
to become
\begin{equation}
-\frac{\hbar^{2}}{2 m_{e}} \frac{d^{2} u(r)}{d r^{2}}+\left[\frac{\hbar^{2} l(l+1)}{2 m_{e} r^{2}}+V(r)\right] u(r)=E u(r)
\end{equation}

where we can set the term in the square brackets to become our effective potential $V_{eff}(r)$, which makes the expression above mathematically equivalent to the Schr√∂dinger equation in one dimension. To solve further we write out our equation in full, inserting the coulomb potential - which we will now need due to its radial dependence.
\begin{equation}
\frac{d^{2} u}{d r^{2}}-\frac{l(l+1)}{r^{2}} u+\frac{2 m_{e} Z e^{2}}{4 \pi \epsilon_{0} \hbar^{2}} \frac{1}{r} u+\frac{2 m_{e} E}{\hbar^{2}} u=0
\end{equation}

For simplicity we make subsitutions of $a_0$, $E_h$ and $\rho$ which are the bohr radius, the Hartree and a unitless coordinate
\begin{equation}
\begin{aligned}
a_{0} &=\frac{4 \pi \epsilon_{0} \hbar^{2}}{m_{e} e^{2}} \\
E_{H} &=\frac{e^{2}}{4 \pi \epsilon_{0} a_{0}} \\
\rho &=Z \frac{r}{a_{0}}
\end{aligned}
\end{equation}

From these substitutions we end up with
\begin{equation}
\frac{d^{2} u}{d \rho}+\left[-\frac{l(l+1)}{\rho^{2}}+\frac{2}{\rho}+\frac{2 E}{Z^{2} E_{H}}\right] u=0
\end{equation}

we seek bound state solutions that are within the atom, so we can impose furth constraints that $E<0$. We can then on to make further shorthand notation substitutions, such that 

\begin{equation}
\beta=2 \sqrt{-2 \frac{E}{Z^{2} E_{H}}}
\end{equation}

and use this to scale our coordinate: $x= \beta \rho$. with these new substitutions we have

\begin{equation}
\left[\frac{d}{d x^{2}}-\frac{l(l+1)}{x^{2}}+\frac{2}{\beta x}-\frac{1}{4}\right] u(x)=0
\end{equation}

Now, lets take a moment to reflect on the physics of the problem. Recall that $x$ is just our scaled
radial coordinate $r$. We want a wavefunction that is finite at $r=0$ and tends to zero as $r \rightarrow \infty$.

As $r \rightarrow \infty$ we find that out that in this limit out equation of motion becomes 

\begin{equation}
\frac{d u(x)}{d^{2} x}=\frac{1}{4} u(x)
\end{equation}

Which has solutions

\begin{equation}
u(x) \sim\left\{\begin{array}{l}
e^{-\frac{x}{2}} \\
e^{+\frac{x}{2}}
\end{array}\right.
\end{equation}

We diitch the second solution and keep the first solution which decays as $r \rightarrow \infty$ - which is the physical property we want from our model. in the limit where $r \rightarrow 0$ (or $x \rightarrow 0$) we find that our equation of motion reduces down to the second term- which dominates. 

\begin{equation}
\frac{d u(x)}{d^{2} x}=\frac{l(l+1)}{x^{2}}
\end{equation}

This has solutions 

\begin{equation}
u(x) \sim\left\{\begin{array}{l}
x^{l+1} \\
x^{-l}
\end{array}\right.
\end{equation}

Turning to the phsyics of the problem again, we find that the second solution if unphysical as it tends to infinity in the limit $x \rightarrow 0$. Thus we can now have a trial somplete solution by combining these two limits! This solution is $u(x)=x^{l+1} e^{-\frac{x}{2}} v(x)$. When placed in our equation of motion this then yeilds

\begin{equation}
x v_{l}^{\prime \prime}(x)+(2 l+2-x) v_{l}^{\prime}(x)+\left(\frac{2}{\beta}-l+1\right) v_{l}(x)=0
\end{equation}

This is the This is Laguerre‚Äôs equation!!! its has been previously solved by brainy mathematicians, meaning that we no longer need to do any more work. The solutions are the Laguerre polynomials

\begin{equation}
v_{l}(x)=L_{k}^{2 l+1}(x)
\end{equation}

and these solutions exist if (and only if) $k$ is a non negative integer, i.e:

\begin{equation}
k=\frac{\beta}{2}-l-1=0,1,2 \ldots
\end{equation}

this sets the bound on the orbital quantum number $l$ such that it number be at least one less than $\beta/2$ which we rename $n$ - the priniciple quantum number

\begin{equation}
l=0,1,2 \ldots n-1
\end{equation}

if we now substitute everything back into our solution and normalise, we can get our entire radial wave function solution in terms of useful parameters.

$$
R_{n l}(r)=N\left(\frac{2 Z r}{n a_{0}}\right)^{l} e^{-\frac{z}{n a_{0}} r} L_{n-l-1}^{2 l+1}\left(\frac{2 Z r}{n a_{0}}\right)
$$
where
$$
N=\left[\left(\frac{2 Z}{n a_{0}}\right)^{3} \frac{(n-l-1) !}{2 n(n+l) !}\right]^{\frac{1}{2}}
$$

the first few solutions from ground state are

\begin{equation}
\begin{array}{l}
R_{10}=\left(\frac{Z}{a_{0}}\right)^{\frac{3}{2}} 2 e^{-\frac{Z r}{a_{0}}} \\
R_{20}=\left(\frac{Z}{2 a_{0}}\right)^{\frac{3}{2}} 2\left(1-\frac{Z r}{2 a_{0}}\right) e^{-\frac{Z r}{2 a_{0}}} \\
R_{21}=\left(\frac{Z}{2 a_{0}}\right)^{\frac{5}{2}} \frac{2}{\sqrt{3}} r e^{-\frac{z r}{2 a_{0}}}
\end{array}
\end{equation}

```python
import matplotlib.cm as cm
import scipy.special as ss
import numpy as np
import math, pdb, sympy

def Laguerre(p,q):
    x = sympy.Symbol('x')
    exp = (-1)**p*sympy.diff(sympy.simplify(sympy.E**x*sympy.diff(sympy.E**(-x)*x**q,x,q)),x,p)
    return lambda x: eval(str(exp))

def SH(l,m,theta,phi):
    return ((-1)**m)*ss.sph_harm(m, l, phi, theta)
    


def SphericalHarmonic(l,m,theta,phi,epsilon):
    
    return epsilon*np.sqrt((2.*l+1)*math.factorial(l-np.abs(m))/(4*np.pi*math.factorial(l+np.abs(m))))*np.real(np.exp(1.j*m*phi)*ss.lpmv(m,l,np.cos(theta)))

def Radial(n,l,r,a):
    p = r/(n*a)
    lag = Laguerre(2*l+1,n+l)
    
    return np.sqrt((2./n/a)**3*math.factorial(n-l-1)/(2.*n*(math.factorial(n+1))**3))*np.exp(-p)*(2*p)**l*lag(2*p)

# def plot_wf_3d(n, l, m, a):
    

def plot_wavefunction(n,l,m,a):
    
    fig = plt.figure(figsize=(10, 10))

    # parameters for the plot (tuned such that the bohr radius is on the same scale as 
    # pixels #)
    resolution = 300; frame_apothem = 300

    # constant for the spherical harmonics (epsilon)
    if m>=0:
        eps = (-1)**m
    else:
        eps = 1

    # create array of data points
    x = np.linspace(-frame_apothem*1.6,frame_apothem*1.6,int(resolution*1.6))
    y = np.linspace(-frame_apothem,frame_apothem,resolution)
    X, Y = np.meshgrid(x,x)
    # create an array of wavefunction values (1e-10 added so that arctan never sees X/0)
    Z = np.abs(Radial(n,l,np.sqrt((X**2+Y**2)),a)*SphericalHarmonic(l,m,np.arctan(X/(Y+1e-10)),0,eps))**2
    Z = Z.astype(np.float)
    #Z = np.sqrt(Z)	# this is done to "raise" the lower, less perceptible values to sight
    # plot the wavefunction in grayscale

    plt.imshow(Z)
    plt.colorbar()
    plt.show()
```

```python
import numpy
import scipy.special
import scipy.misc
from enthought.mayavi import mlab

r = lambda x,y,z: numpy.sqrt(x**2+y**2+z**2)
theta = lambda x,y,z: numpy.arccos(z/r(x,y,z))
phi = lambda x,y,z: numpy.arctan(y/x)
#phi = lambda x,y,z: numpy.pi+numpy.select(
#	[x>0, x==0, x<0],
#	[
#		numpy.arctan(y/x),
#		.5*numpy.pi*numpy.sign(y),
#		numpy.arctan(y/x)+numpy.pi*numpy.sign(y)]
#)
a0 = 1.
R = lambda r,n,l: (2*r/n/a0)**l * numpy.exp(-r/n/a0) * scipy.special.genlaguerre(n-l-1,2*l+1)(2*r/n/a0)
WF = lambda r,theta,phi,n,l,m: R(r,n,l) * scipy.special.sph_harm(m,l,phi,theta)
absWF = lambda r,theta,phi,n,l,m: abs(WF(r,theta,phi,n,l,m))**2

x,y,z = numpy.ogrid[-24:24:55j,-24:24:55j,-24:24:55j]

mlab.figure()

#mask = numpy.select([theta(x,y,z)>numpy.pi/3.],[numpy.select([abs(phi(x,y,z))<numpy.pi/3.],[numpy.nan],default=1)],default=1)
mask = 1

for n in range(2,3):
	for l in range(1,n):
		for m in range(-l,l+1,1):
			w = absWF(r(x,y,z),theta(x,y,z),phi(x,y,z),n,l,m)
			mlab.contour3d(w*mask,contours=6,transparent=True)

mlab.colorbar()
mlab.outline()
mlab.show()
```
