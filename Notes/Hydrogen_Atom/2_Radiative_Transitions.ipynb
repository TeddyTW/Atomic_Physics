## Radiative Transitions of Hydrogen

The solutions we have found to the Schrödinger equation are eigenstates and are constant in time. These states are referred to as stationary states, as they are solutions to the time independent Schrödinger equation, which  results from making the Potential $V(r, t)$ constant in time. These solutions are constant in time, and to make transitions between them, we require a time varying potential - or an interaction. A time varying potential large enough to induce transitions can come from small pertubations from an oscillating electric field (light!) 

### Fermi's Golden Rule: TIPT derivation

For interactions with light, we want to study how quantum systems in stationary states respond to small time dependent perturbations (light) and how they settle into stationary states after. we intende to study the transitions induced by a perturbation between stationary states of the unperturbed system. To do that we introduce a hamiltonian of two parts: 

\begin{equation}
    H = H_0 + H_I(t)
\end{equation}

The first part is a Stationary part $H_0$ and the second is the small time dependent interaction part $H_I(t)$, which is a result of the electric field oscillation. Since $H_0$ is the unperturbed hamiltonian, the time independent Schrodinger
equation is still satisfied by this part:

\begin{equation}
H_{0}\left|\varphi_{n}\right\rangle=E_{n}\left|\varphi_{n}\right\rangle
\end{equation}

Where our "total" wavefunction (including time dependence) is given by:

\begin{equation}
\left|\psi_{n}(t)\right\rangle=\left|\varphi_{n}\right\rangle e^{-i E_{n} t / \hbar}
\end{equation}

The total Hamiltonian $H = H_0 + H_I(t)$ satisfies the Time dependent schrodinger equation:
\begin{equation}
H|\psi(t)\rangle=\left[H_{0}+H_I(t)\right]|\psi(t)\rangle=i \hbar \frac{\partial|\psi(t)\rangle}{\partial t}
\end{equation}

Using the completeness of the set of stationary state wavefunctions, we can define any wavefunction as a linear combination of these states. Therefore we can define:
\begin{equation}
|\psi(t)\rangle=\sum_{n} c_{n}(t)\left|\psi_{n}(t)\right\rangle=\sum_{n} c_{n}(t)\left|\varphi_{n}\right\rangle e^{-i E_{n} t / \hbar}
\end{equation}

If we then insert this into our Time dependent Schrodinger equation:

\begin{equation}
\begin{array}{c}
H_{0} \sum_{k} c_{k}(t)\left|\varphi_{k}\right\rangle e^{-i E_{k} t / \hbar}+H_I(t) \sum_{k} c_{k}(t)\left|\varphi_{k}\right\rangle e^{-i E_{k} t / \hbar} \\
=i \hbar \frac{\partial}{\partial t} \sum_{k} c_{k}(t)\left|\varphi_{k}\right\rangle e^{-i E_{k} t / \hbar} \\
\\
\sum_{k} c_{k}(t) E_{k}\left|\varphi_{k}\right\rangle e^{-i E_{k} t / \hbar}+\sum_{k} c_{k}(t) H_I(t)\left|\varphi_{k}\right\rangle e^{-i E_{k} t / \hbar} \\
=i \hbar \sum_{k} \frac{\partial c_{k}(t)}{\partial t}\left|\varphi_{k}\right\rangle e^{-i E_{k} t / \hbar}+i \hbar \sum_{k} c_{k}(t)\left|\varphi_{k}\right\rangle\left(-\frac{i E_{k}}{\hbar}\right) e^{-i E_{k} t / \hbar}\\
\\
\end{array}
\end{equation}

We can then close with a $\langle \varphi_n | $ 

\begin{equation}
\begin{array}{c}
\sum_{k} c_{k}(t) E_{k}\langle \varphi_n \left|\varphi_{k}\right\rangle e^{-i E_{k} t / \hbar}+\sum_{k} c_{k}(t) \langle \varphi_n | H_I(t)\left|\varphi_{k}\right\rangle e^{-i E_{k} t / \hbar}
=i \hbar \sum_{k} \frac{\partial c_{k}(t)}{\partial t}\langle \varphi_n \left|\varphi_{k}\right\rangle e^{-i E_{k} t / \hbar}+i \hbar \sum_{k} c_{k}(t)\langle \varphi_n \left|\varphi_{k}\right\rangle\left(-\frac{i E_{k}}{\hbar}\right) e^{-i E_{k} t / \hbar}\\
\\
c_{n}(t) E_{n} e^{-i E_{n} t / \hbar}+\sum_{k} c_{k}(t) \langle \varphi_n | H_I(t)\left|\varphi_{k}\right\rangle e^{-i E_{k} t / \hbar}
=i \hbar \frac{\partial c_{n}(t)}{\partial t} e^{-i E_{n} t / \hbar} + i \hbar c_{n}(t)\left(-\frac{i E_{n}}{\hbar}\right) e^{-i E_{n} t / \hbar}\\
\\
\sum_{k} c_{k}(t) \langle \varphi_n | H_I(t)\left|\varphi_{k}\right\rangle e^{-i E_{k} t / \hbar}
=i \hbar \frac{\partial c_{n}(t)}{\partial t} e^{-i E_{n} t / \hbar} \\
\\
\sum_{k} c_{k}(t) H_{nk} e^{-i E_{k} t / \hbar}
=i \hbar \frac{\partial c_{n}(t)}{\partial t} e^{-i E_{n} t / \hbar} \\
\\
\frac{\partial c_{n}(t)}{\partial t}=\frac{1}{i \hbar} \sum_{k} c_{k}(t) H_{n k}(t) e^{i \omega_{n k} t} \quad \text{where} \quad \omega_{n k}=\left(E_{n}-E_{k}\right) / \hbar
\end{array}
\end{equation}

Up to this point we have made no approximation. The difficulty in solving the above equation stems from the coefficients being expressed in terms of themselves. In order to evaluate the coefficients we make two assumptions:

1. The system is initially in state $|i>$, thus, all of the coefficients at $t = 0$ are equal to zero, except for $c_i$
    
2. The perturbation is very weak and applied for a short period of time, such that all of the coefficients remain nearly unchanged.

Therefore, we get:

\begin{equation}
\frac{\partial c_{n}(t)}{\partial t}=\frac{1}{i \hbar} c_{i}(t) H_{n i}(t) e^{i \omega_{n i} t}
\end{equation}

for any final state $<f|$ we find that $c_f(t) \approx c_f(0) = 0$ therefore:

\begin{equation}
c_{f}(t)=\frac{1}{i \hbar} \int_{0}^{t} H_{f i}\left(t^{\prime}\right) e^{i \omega_{f i} t^{\prime}} d t^{\prime}
\end{equation}

The above equation demonstrates that a small pertubation to the hamiltonian $H_I$ can induce transitions from some initial state $i$ to some other final state $f$. So the probability of the wavefunction $|\Psi>$ undergoing this transition $i \rightarrow f$ is $|c_f|^2$. This is due to the completeness of $|\Psi>$ is equal to $<\phi_f|\Psi>$.

## Radiative Transitions

The solutions which we have found to the Hydrogen Stomic Schrodinger equation eigenstates which are constant in time and are refered to as staitonary states. An interaction is required to transition between stationary state, which we will conider in the form of an electromagnetic field. This is a semiclassical approach, as a fulk quantum approach would require the quantistaion of the EM field into Photons. Consider the osccilating electromagentic field with polarisation $\hat{e}_{rad}$.

\begin{equation}
\vec{E}=E_{0} \operatorname{Re}\left\{e^{-i \omega t} \hat{e}_{rad}\right\}
\end{equation}

Time-dependent perturbation theory is used to determine the probability associated with a transition from intiial state $i$ to final sate $f$ as being given by Fermis Golden Rule.

\begin{equation}
\Gamma=C\left|\int \psi_{j}^{*}(r) \widehat{H}_{I} \psi_{i}(r) d^{3} r\right|^{2}=C\left|\left\langle\psi_{j}\left|\widehat{H}_{I}\right| \psi_{i}\right\rangle\right|^{2}=C\left|\left\langle j\left|\widehat{H}_{I}\right| i\right\rangle\right|^{2}
\end{equation}

For which $\Gamma$ is a rate, with units $s^{-1}$. This a completely general expression to which we can apply any interaction hamiltonian. The interaction hamiltonian of our osicallating electic field is $\widehat{H}_{I}(t)=e \vec{r} \cdot \vec{E}(t)$ where $-e\vec{r}$ is our dipole operator. This is a dipole approximation of the full interation hamiltonian $\widehat{H}_{I}=\frac{e}{2 m}(\vec{p} \cdot \vec{A}-\vec{A} \cdot \vec{p})+\frac{e^{2}}{2 m} \vec{A}^{2}-e \phi$, which is valid under the assumption that the wavelength is long compared to the size of the dipole $\lambda >> a_0$. Thus we end up with

\begin{equation}
\Gamma=C e^{2} E_{0}^{2}\left|\left\langle j\left|\vec{r} \cdot \hat{e}_{r a d}\right| i\right\rangle\right|^{2}
\end{equation}

Where the bra/ket term can be seperated into the angular and radial parts as follows

\begin{equation}
\left\langle j\left|\vec{r} \cdot \hat{e}_{r a d}\right| i\right\rangle=D_{i j} I_{a n g}
\end{equation}

where
\begin{equation}
D_{i j}=\int_{0}^{\infty} R_{n j l_{j}}^{*}(r) r R_{n_{i} l_{i}}(r) r^{2} d r\\
I_{a n g}=\iint_{0}^{2 \pi \pi}\left(Y_{l_{j}}^{m_{l_{j}}}(\theta, \phi)\right)^{*} \hat{r} \cdot \hat{e}_{r a d} Y_{l_{i}}^{m_{l_{i}}}(\theta, \phi) \sin \theta d \theta d \phi
\end{equation}

Interestingly we find that $D_{ij}=D_{ji}$ meaning that the probablility of the transiton is the same irrespective of its direction (increasing or decreasing in energy). To investigate how $I_{ang}$ behaves we need a to play around with the mathematics a bit. Let us consider the vectors $\vec{r}$ and $\hat{e}_{r a d}$ First Lets consider the polarisation vector $\hat{e}_{r a d}$, it can be written in terms of its $z$ component (amplitude $A_\pi$) and its component in the $xy$-plane, written as a superposition of two circular polarisations (amplitudes $A_{\sigma^+}$, $A_{\sigma^-}$)

\begin{equation}
\widehat{\mathrm{e}}_{\mathrm{rad}}=A_{\sigma^{-}}\left(\frac{\hat{\mathrm{e}}_{x}-\mathrm{i} \hat{\mathrm{e}}_{y}}{\sqrt{2}}\right)+A_{\pi} \hat{\mathrm{e}}_{z}+A_{\sigma^{+}}\left(-\frac{\hat{\mathrm{e}}_{x}+\mathrm{i} \hat{\mathrm{e}}_{y}}{\sqrt{2}}\right)
\end{equation}

We then can rewrite $\vec{r}$ in spherical polars:


\begin{equation}
\begin{aligned}
\hat{r} &=\frac{1}{r}(x\hat{\mathrm{e}}_{x}+y\hat{\mathrm{e}}_{y}+z \hat{\mathrm{e}}_{z}) \\
&=\sin \theta \cos \phi\hat{\mathrm{e}}_{x}+\sin \theta \sin \phi \hat{\mathrm{e}}_{y}+\cos \theta\hat{\mathrm{e}}_{z}
\end{aligned}
\end{equation}

which can then be further written in terms of spherical harmonics, which gives:

\begin{equation}
\hat{r}=\sqrt{\frac{4 \pi}{3}}\left[\frac{1}{\sqrt{2}}(\hat{\mathrm{e}}_{x}+i \hat{\mathrm{e}}_{y}) Y_{1}^{-1}+\frac{1}{\sqrt{2}}(-\hat{\mathrm{e}}_{x}+i \hat{\mathrm{e}}_{y}) Y_{1}^{1}+Y_{1}^{0} \hat{\mathrm{e}}_{z}\right]
\end{equation}

We can now write the dot-product in terms of the spherical harmonics:
\begin{equation}
\hat{r} \cdot \hat{e}_{r a d}=\sqrt{\frac{4 \pi}{3}}\left(a_{\sigma^{-}} Y_{1}^{-1}+a_{\sigma^{+}} Y_{1}^{1}+a_{\pi} Y_{1}^{0}\right) .
\end{equation}

## Selection Rules
This all seems a bit abritary, but there is some insight coming.

### Linear Polarisation
Consider the component of the electric field parallel to the z-axis $A_z$, which will induce a dipole moment on the atom proportional to $\hat{e}_{rad} \cdot \hat{e}_z = \cos{\theta}$ making the integral over the angular parts of the wavefunction
\begin{equation}
  \mathcal{I}_{\mathrm{ang}}^{\pi}=\int_{0}^{2 \pi} \int_{0}^{\pi} Y_{l_{2}, m_{2}}^{*}(\theta, \phi) \cos \theta Y_{l_{1}, m_{1}}(\theta, \phi) \sin \theta \mathrm{d} \theta \mathrm{d} \phi  
\end{equation}
The probability of transition should be unchanged by any cylindrical rotation of the system, and thus $\mathcal{I}_{\mathrm{ang}}^{\pi}$ is expected to demonstrate cylindrical symmetry, therefore $\mathcal{I}_{\mathrm{ang}}^{\pi}=\mathrm{e}^{\mathrm{i}\left(m_{1}-m_{2}\right) \phi_{0}} \mathcal{I}_{\mathrm{ang}}^{\pi}$, thus leading to the selection $\Delta m_l = 0$. 

### Circular Polarisation
We do the same with Circularly Polarised light, the angular intergral comes to

\begin{equation}
I_{a n g} \sim \iint_{0}^{2 \pi \pi}\left(Y_{l_{j}}^{m_{l_{j}}}\right)^{*} \sin \theta e^{\pm i \phi} Y_{l_{i}}^{m_{l_{i}}} \sin \theta d \theta d \phi
\end{equation}

Again, we consider the of symmetry with respect to rotation about the zaxis through an arbitrary angle.  $\mathcal{I}_{\mathrm{ang}}^{\pi}=\mathrm{e}^{\mathrm{i}\left(m_{1}-m_{2} \pm 1 \right) \phi_{0}} \mathcal{I}_{\mathrm{ang}}^{\pi}$ which (using the same logic as previously) states that either $I_{ang}=0$ or that  $\Delta m_l = \pm 1$

### Parity
To get the $l$ selection rules, we must consider the parity transformation, which transforms the wavefunction to its mirror image $r \rightarrow -r$, equivalent to the simultaneous transformations $\theta \rightarrow \pi - \theta$ and $\phi \rightarrow \phi + \pi$. The hydrogen atom should have the same energy levels after a parity transformation as the Coulomb potential is the same after reflection and electromagnetic forces have $P$ symmetry. The parity operator $\widehat{P}$ returns the parity $P$ of a wavefunction $\widehat{P}\psi= P\psi$ where $P$ which is 1 or $-1$ for even and odd wavefunctions respectively. By looking at the spherical harmonics of $Y$ it is fairly straightforward to determine that $\widehat{P}Y_{l,m}= (-1)^lY_{l_m}$.
For the probability of transitions to be invariant under parity transformations, respecting $P$ symmetry, $\mathcal{I}_{ang}$ must also be invariant, i.e
\begin{equation}
    \widehat{P}\mathcal{I}_{\text {ang }}=(-1)^{l_1 + l_2 + 1}\mathcal{I}_{\text {ang }} = \mathcal{I}_{\text {ang }}
\end{equation}
For the above statement to be true, either $\mathcal{I}_{\text {ang }} = 0$ (forbidden transition) or $l_2 - l_1 = \Delta l = \pm 1$. 

With a bit of intuition we can determine that a singular photon carries $\pm \hbar$ or angular momentum, which is transfered to the Atom when it interacts with the electron. Thus, the selection rule can be seen as conserving angular momentum

## Summary of Selection Rules

\begin{equation}
\begin{array}{|c|l|}
\hline \text { Selection rule } & {\text { Interpretation }} \\
\hline \Delta m_{l}=\pm 1 & \text { Circular polarisation } \\
\hline \Delta m_{l}=0 & \text { Linear polarisation } \\
\hline \Delta l=\pm 1 & \text { Conservation of angular momentum } \\
\hline \Delta l \neq 0 & \text { Conservation of parity } \\
\hline
\end{array}
\end{equation}
