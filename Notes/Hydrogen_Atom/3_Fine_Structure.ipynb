We study the real hydrogen atom and discover that the coupling of the electron spin with orbital angular momentum leads to a small but clearly observable splitting in the lines in the spectrum of hydrogen. 

Electrons have spin. Spin is a degree of freedom with conservation laws in quantum mechanics which resemble angular momentum, but there is no adequate classical analogue for spin. Despite the name there is no motion associated with spin, instad the elctrons apear to have a dicreeet intrinsic angualr momentum not associated with any specific motion. The eigenstates of the spin operator and its projection along the ùëß-axis are given by:

\begin{equation}
\begin{array}{l}
\vec{s}^{2}\left| s, m_{s}\right\rangle=\hbar^{2} s(s+1)\left| s, m_{s}\right\rangle \\
\hat{s}_{z}\left| s, m_{s}\right\rangle=\hbar m_{s}\left| s, m_{s}\right\rangle
\end{array}
\end{equation}

Spin obeys the same commutation rules as angular momentum $\left[s_{x}, s_{y}\right]=i \hbar s_{z}$ so exactly the same formalism applies. So for example, if we have a state with a siven $s$ it means that $m_s$ can take values $m_s = \{-s, -s+1, ... 1-s, s\}$. For a single electron $s$ is $\frac{1}{2}$ which means $m_s = +\frac{1}{2} \quad \text{or} \quad -\frac{1}{2}$

\begin{equation}
s_{z}\left|\frac{1}{2}\left(\pm \frac{1}{2}\right)\right\rangle=\pm \frac{\hbar}{2}\left|\frac{1}{2}\left(\pm \frac{1}{2}\right)\right\rangle
\end{equation}

The $z$ component states $m_s = \pm \frac{1}{2}$ are often labeled as $|\uparrow\rangle$ and $|\downarrow\rangle$ repectively. The electron has an intrinsic magnetic moment due to its spin, which given by

\begin{equation}
\vec{\mu}=-g_{s} \mu_{B} \hat{\vec{s}}
\end{equation}

These new quatum number are now essential for the decription of the state of the hydrogen wavefunction. We need five total quantum numbers to specify the state of the hydrogen atom, thus, the wavefunction becomes:

\begin{equation}
|\psi\rangle=\left|n l m_{l} s m_{s}\right\rangle
\end{equation}

Spin arises in a natural way in the from the relativistic Dirac equation. We will study some of the consequences arising from a non-relativistic expansion of the full relativistic model, rather than full Dirac theory

## The relativistic-mass correction

Total energy is given by
\begin{equation}
\widehat{H}=\left(p^{2} c^{2}+m_{0}^{2} c^{4}\right)^{\frac{1}{2}}+V(r)
\end{equation}

$m_0$ is the rest mass of the electron. So we can derive the relativistic kinetic energy from an expansion

\begin{equation}
\begin{aligned}
&K=\left(p^{2} c^{2}+m_{0}^{2} c^{4}\right)^{\frac{1}{2}}-m_{0} c^{2} \\
&=m_{0} c^{2}\left(1+\frac{p^{2}}{m_{0}^{2} c^{2}}\right)^{\frac{1}{2}}-m_{0} c^{2} \\
&\simeq \frac{p^{2}}{2 m_{0}}-\frac{1}{8} \frac{p^{4}}{m_{0}^{3} c^{2}}+\cdots \\
&=K_{0}-\frac{1}{8} \frac{p^{4}}{m_{0}^{3} c^{2}} \\
&=K_{0}-\frac{1}{2 m_{0} c^{2}}\left(\frac{p^{2}}{2 m_{0}}\right)^{2} \\
&=K_{0}-\frac{1}{2 m_{0} c^{2}} K_{0}^{2}
\end{aligned}
\end{equation}

So we end up with a low order correction to the non-relativistic kinetic energy.

\begin{equation}
\begin{aligned}
\Delta K &=-\frac{1}{2 m_{0} c^{2}} K_{0}^{2} \\
&=-\frac{1}{2 m_{0} c^{2}}[E-V(r)]^{2} .
\end{aligned}
\end{equation}

Thus we can now apply this first order pertubation to the energy levels of the hydrogen wavefunction:

\begin{equation}
\begin{aligned}
\Delta E_{n}^{\prime} &=-\frac{1}{2 m_{0} c^{2}}\left\langle[E-V(r)]^{2}\right\rangle \\
&=-\frac{1}{2 m_{0} c^{2}}\left[E_{n}^{2}-2 E_{n}\left\langle-\frac{Z e^{2}}{4 \pi \epsilon_{0}} \frac{1}{r}\right\rangle+\left\langle\frac{Z^{2} e^{4}}{\left(4 \pi \epsilon_{0}\right)^{2}} \frac{1}{r^{2}}\right\rangle\right]
\end{aligned}
\end{equation}

We already know the energy levels $E_n$ and we already have the results for the expectation values so collecting it all and writing it in terms of the fine structure constant, the relativistic-mass correction becomes

\begin{equation}
\Delta E_{n}^{\prime}=-\frac{\alpha^{2} Z^{2}}{n^{2}} E_{n}\left(\frac{3}{4}-\frac{n}{l+\frac{1}{2}}\right)
\end{equation}

The correction depends on both the principle atomic number $n$ and the orbital atomic number $l$

Again, our new operator has the same algebra as $l$ and $s$, with eigenstates and eigenvalues:

\end{equation}

## Spin orbit interaction

### $\mathbf{B}$ field experienced by orbiting electron
An electron moving through an electric field $\mathbf{E}$ experiences an effective magnetic field $\mathbf{B}$ due to the way an electric field behaves under a Lorentz transformation from a stationary to a moving frame in special relativity.

\begin{equation}
\mathbf{B}=-\frac{1}{c^{2}} \mathbf{v} \times \mathbf{E}
\end{equation}

We can substitute for the electric field in terms of the gradient of the potential energy $V$ and unit vector in the radial direction, this is the form more useful in terms of the Schrodinger equation. Using the fact that the electrons potential energy is equal to the electrostatic potential multiplied by the electrons charge $-e$. Thus, we get expression:

\begin{equation}
\mathbf{E}=\frac{1}{e} \frac{\partial V}{\partial r} \hat{e_r}
\end{equation}

We can then substitute this in. Considering that orbital angular momentum $l$ is equal to $\hbar \mathbf{l}=\mathbf{r} \times m_{\mathrm{e}} \mathbf{v}$, with a bit of rearrangement get an expression for the magnetic field $\mathbf{B}$ experienced by the electron:

\begin{equation}
\mathbf{B}=\frac{1}{m_{\mathrm{e}} c^{2}}\left(\frac{1}{e r} \frac{\partial V}{\partial r}\right) \mathbf{r} \times m_{\mathrm{e}} \mathbf{v}=\frac{\hbar}{m_{\mathrm{e}} c^{2}}\left(\frac{1}{e r} \frac{\partial V}{\partial r}\right) l
\end{equation}

### Electron magnetic moment $\mathbf{\mu}$

If the electron is visualized as a classical charged particle rotating about an axis with angular momentum $\vec{L}$, its magnetic dipole moment $\mu$ is given by:

\begin{equation}
\boldsymbol{\mu}=\frac{-e}{2 m_{\mathrm{e}}} \vec{L}
\end{equation}

Now, we want to switch the the spin angualr momentum $\vec{S}$, which is intrinsic to the electron. This also introduces a proportionality factor, $g_s \approx 2$, called the g-factor. We also want to express our magnetic moment in units of $\mu_B$- the Bohr magneton $\mu_{\mathrm{B}}=\frac{e \hbar}{2 m_{\mathrm{e}}}$, corresponding to a spin expressed in units of $\hbar$. Thus, we get

\begin{equation}
\boldsymbol{\mu}_{\mathrm{s}}=-g_{\mathrm{s}} \frac{e \hbar}{2 m_{\mathrm{e}}} \frac{\vec{S}}{\hbar} = -g_{\mathrm{s}} \frac{e \hbar}{2 m_{\mathrm{e}}} \mathbf{s}
\end{equation}

### Shift in electron energy
The energy of a magnetic moment in a magnetic field is given by

\begin{equation}
\Delta H=-\boldsymbol{\mu} \cdot \mathbf{B}
\end{equation}

Thus, we can expect our electrons energy to shift accordingly, due to the energy of the electrons spin magnetic moment $\mu_s$ in the √´xternal"field produced by the electrons angular momentum in the rest frame of the electron.

\begin{equation}
\widehat{H}_{s o}=\frac{\hbar^{2}}{2 m_{0}^{2} c^{2}} \frac{1}{r} \frac{d V}{d r} \vec{s} \cdot \vec{l}
\end{equation}

Taking into account the Thomas Precession, a relativistic correction that applies to the spin of an elementry particle following a curvilinear orbit, we have a second correction to the g-factor. The end result is:

\begin{equation}
H_{\mathrm{s}-\mathrm{o}}=\left(g_{s}-1\right) \frac{\hbar^{2}}{2 m_{e}^{2} c^{2}}\left(\frac{1}{r} \frac{\partial V}{\partial r}\right) \mathbf{s} \cdot 1
\end{equation}

The expectation value of this Hamiltonian gives us the energy change. Substituing in our coulomb potential for hydrogen $\frac{1}{r} \frac{\partial V}{\partial r}=\frac{Z^2e^{2} / 4 \pi \epsilon_{0}}{r^{3}}$, and assuming approximation $g_s - 1 = 1 $ we get:

\begin{equation}
E_{\mathrm{s}-\mathrm{o}}=\frac{\hbar^{2}}{2 m_{\mathrm{e}}^{2} c^{2}} \frac{e^{2}}{4 \pi \epsilon_{0}}\left\langle\frac{1}{r^{3}}\right\rangle\langle\mathbf{s} \cdot 1\rangle
\end{equation}

We know from previous calculations what the value of $\left\langle\frac{1}{r^{3}}\right\rangle$ is:

\begin{equation}
\left\langle\frac{1}{r^{3}}\right\rangle=\frac{1}{l\left(l+\frac{1}{2}\right)(l+1) n^{3}}\left(\frac{Z}{a_{0}}\right)^{3}
\end{equation}

We also want to substitute our expression for energy level (derived in previous sections) for brevity, which I will reiterate here:

\begin{equation}
E_{n}=-\frac{Z^{2} e ^2}{8 \pi \epsilon_0 a_0} \frac{1}{n^{2}}
\end{equation}

Thus, using the fine structure constant $\alpha=\frac{1}{4 \pi \varepsilon_{0}} \frac{e^{2}}{\hbar c}$,  and our value for bohr radius $a_{0}=\frac{\hbar^{2}}{\left(e^{2} / 4 \pi \epsilon_{0}\right) m_{e}}$ we get (after a bit of reformatting) a concise formula:

\begin{equation}
\Delta E^{\prime \prime}=-\frac{\alpha^{2} Z^{2}}{n^{2}} E_{n} \frac{n}{l\left(l+\frac{1}{2}\right)(l+1)}\langle\hat{s} \cdot \vec{l}\rangle
\end{equation}

But what about the $\langle\hat{s} \cdot \vec{l}\rangle$ term? How do we deal with this? let us use the fact that $\vec{j} = \vec{l} +\vec{s}$ 
\begin{equation}
\begin{aligned}
&\hat{\vec{j}}^{2}=(\hat{\vec{l}}+\hat{\vec{s}}) \cdot(\hat{\vec{l}}+\hat{\vec{s}})=\hat{\vec{l}}^{2}+\hat{\vec{s}}^{2}+\hat{\vec{s}} \cdot \hat{\vec{l}}+\vec{l} \cdot \vec{s} \\
&=\hat{\vec{l}}^{2}+\hat{\vec{s}}^{2}+2 \hat{\vec{s}} \cdot \hat{\vec{l}},
\end{aligned}
\end{equation}

Therefore

\begin{equation}
\langle\hat{\vec{s}} \cdot \hat{\vec{l}}\rangle=\frac{1}{2}\left\langle\hat{\vec{j}}^{2}-\hat{\vec{l}}^{2}-\hat{\vec{s}}^{2}\right\rangle=\frac{1}{2}[(j(j+1)-l(l+1)-s(s+1)]
\end{equation}

After a slight bit of algebra, we can get a consise expression for $\Delta E ' + \Delta E ''$, taking $j=l+\frac{1}{2}$ and $s=\frac{1}{2}$

\begin{equation}
\Delta E^{\prime}+\Delta E^{\prime \prime}=-\frac{\alpha^{2} Z^{2}}{n^{2}} E_{n}\left[\frac{3}{4}-\frac{n}{j+\frac{1}{2}}\right]
\end{equation}


#### Darwin Term
There is a third correction to the energy levels of Hydrogen that is important to know about- however this falls out  of a treatment of a non-relativistic expansion of the (otherwise relativistic) Dirac equation. I intent to make a page on this expansion at some point, to better understnad where this term comes from- but for now we will shall consider it to be a $\textbf{magical}$ correction to the $s$ orbitals of hydrogen due to its overlap with the nucleus. This correction if only for $s$ orbitals, as other orbitals have $l>0$ and thus do not overlap with the nucleus. For $l=0$ we have $\Psi(r=0)$ not equal to zero, thus we get an extra shift in energy level given by

\begin{equation}
\Delta E^{\prime \prime \prime}=\frac{\pi \hbar^{2}}{2 m_{0}^{2} c^{2}} \frac{Z e^{2}}{4 \pi \epsilon_{0}}|\psi(0)|^{2}
\end{equation}

This works out as

\begin{equation}
\Delta E^{\prime \prime \prime}=-\frac{\alpha^{2} Z^{2}}{n^{2}} E_{n} n
\end{equation}

### Fine Structure
The three correction terms can be combined to give the total change in energy level. When combining these shifts it is important to consider that the Darwin Term is only applicable for $l=0$ and the Spin orbit interaction is only applicable for $l \neq 0$. Thus we get:

\begin{equation}
\Delta E_{F S}=\Delta E^{\prime}+\Delta E^{\prime \prime}+\Delta E^{\prime \prime \prime}=-\frac{\alpha^{2} Z^{2}}{n^{2}} E_{n}\left(\frac{3}{4}-\frac{n}{j+\frac{1}{2}}\right)
\end{equation}
